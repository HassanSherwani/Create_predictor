## 2)- Loading data-files

```python
df = pd.read_csv('testset3.csv')
```

```python
df.head(2)
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df.shape
```

# 3)- Model Preparation

```python
# Selecting all columns except last one that is 'm25'.

X = df.iloc[:,:-1].values          
y = df['m25'].values
```

```python
X
```

```python
X.shape
```

```python
print(type(X))
print(type(y))
```

### normalize features

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled = scaler.fit_transform(X)
```

### Split Data to Train and Test

```python
# As it is neural network so, we tend to use capital letters for train & test sample
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, y, test_size=0.2)
```

```python
print(type(X_Train))
print(type(X_Test))
print(type(Y_Train))
print(type(Y_Test))
```

```python
print(X_Train.shape)
print(X_Test.shape)
print(Y_Train.shape)
print(Y_Test.shape)
```

### reshape input

```python
# reshape input to be 3D [samples, timesteps, features]
X_Train = X_Train.reshape((X_Train.shape[0], 1, X_Train.shape[1]))
X_Test = X_Test.reshape((X_Test.shape[0], 1, X_Test.shape[1]))
```

```python
print(X_Train.shape)
print(X_Test.shape)
print(Y_Train.shape)
print(Y_Test.shape)
```

# 4)-Building Neural Network

```python
# network architecture
model = Sequential()
model.add(LSTM(50, input_shape=(X_Train.shape[1], X_Train.shape[2])))
model.add(Dense(1))
model.compile(loss='mse', optimizer='adam')
```

```python
model.summary()
```

### Fit model

```python
history = model.fit(X_Train, Y_Train, epochs=70, batch_size=10, validation_data=(X_Test, Y_Test), verbose=2, shuffle=False)
```

# 5)-Evaluation

```python
print(X_Train.shape)
print(X_Test.shape)
print(Y_Train.shape)
print(Y_Test.shape)
```

```python
# make a prediction
yhat = model.predict(X_Test)
# Convert X_Test back to natural state 
X_Test = X_Test.reshape((X_Test.shape[0], X_Test.shape[2]))
```

```python
X_Test.shape
```

```python
yhat[:5]
```

```python
# invert scaling for forecast
inv_yhat = concatenate((yhat, X_Test[:, 1:]), axis=1)
inv_yhat = scaler.inverse_transform(inv_yhat)
inv_yhat = inv_yhat[:,0]
# invert scaling for actual
Y_Test = Y_Test.reshape((len(Y_Test), 1))
inv_y = concatenate((Y_Test, X_Test[:, 1:]), axis=1)
inv_y = scaler.inverse_transform(inv_y)
inv_y = inv_y[:,0]
```

```python
# calculate RMSE
from sklearn.metrics import mean_squared_log_error,mean_squared_error, r2_score,mean_absolute_error 

mse = mean_squared_error(inv_y, inv_yhat)
mae = mean_absolute_error(inv_y, inv_yhat)
rmse = sqrt(mean_squared_error(inv_y, inv_yhat))
r2 = r2_score(inv_y, inv_yhat)
```

```python
print('')
print('MSE    : %0.2f ' % mse)
print('MAE    : %0.2f ' % mae)
print('RMSE   : %0.2f ' % rmse)
print('R2     : %0.4f ' % r2)
```
